# Feature enginnering :- 
- when i was doing the data cleaning in that particular file i have done feature engineering 
- column name was family_size


```python
def family_type(number):
    if number==0:
        return "Alone"
    elif number>0 and number<=4:
        return "Medium"
    else:
        return "Large"
```

```python
df["family_size"] = df["family_size"].apply(family_type)
```

```python
pd.crosstab(df["family_size"] ,df["Survived"]).apply(lambda r :round(( r/r.sum())*100,1) , axis =1)
```

# Detecting outliers


```python
# handling ourliers in age (Almost normal)
df = df[df["Age"] <(df["Age"].mean() + 3*df["Age"].std())]
```

```python
df.shape
```

```python
# Handling outliers from the fare column 
Q1 = np.percentile(df["Fare"] ,25)
Q3 = np.percentile(df["Fare"] ,75)

outlier_low = Q1 -1.5*(Q3-Q1)
outlier_hight = Q3 +1.5 *(Q3-Q1)

df = df[(df['Fare']>outlier_low) & (df["Fare"] <outlier_hight)]
```

```python
# One hot encoding 
df.sample(4)
```

```python
df2 = pd.get_dummies(data = df, columns=["Pclass" ,"Sex" ,"Embarked" ,"family_size"] , drop_first=True)
```

```python
sns.heatmap(df.corr() ,cmap = "summer")
```

# Conclusion 
- Chances of female survival is higher than the male survival

- Somehow , peoples going to C survived more
- Peoples range in between 20 and 40 are having high chance to not survive
- Peoples travelling with smaller families are having highest chance to survive as compare to peoples with large families and travelling alone
